```python
#RGB -RED,GREEN,BLUE min=0 max=255 
#discoloring image using  numpy
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
imag = Image.open(r"C:\Users\Dell\mountain.jpg")
img= np.array(imag)
img[:,:,0]=np.clip(img[:,:,0]+155,100,1555)
plt.imshow(img)
plt.axis("off")
plt.show()
```
