```python
c=np.array([[2,2,2],[1,1,1]])
print(b+c)
print(b-c)
print(b*c) #not applicable as it dont follow matrix princple if order 2*3 then 2*3=3*2 to get correct result take transpose
print(b/c) 
print(b**c)
print(b//c)
```

```python
d=np.transpose(c)
print(d)
```

```python
print(np.dot(b,d))#use dot product to perform mulitplication of 2 d array
```
