```python
#fibonacci series
a=list(map(int,input("enter a number separated by sppaces").split()))
b=len(a)
for i in range(2,len(a)):
    if a[i]==a[i-1]+a[i-2]:
        continue
    else:
        print("not in fibonacci series")
        break
else: print("fibonacci series")
```

```python
#palindrome
string=input("enter a string")
s=list(string)
for i in range(len(s)):
    if s[i]==s[len(s)-i-1]:
        continue
    else: 
        print("not palindrome")
        break
else:print("palindrome")
```

```python
#basic calculator
n1=float(input("enter a number1: "))
n2=float(input("enter a number2: "))
while True:
    print("\n 1.ADDITION \n 2.SUBTRACTION \n 3.MULTIPLICATION \n 4.DIVISION \n 5.FLOOR DIVISION \n 6.TO GET REMAINDER\n 7.EXIT \n")
    o=int(input("enter you choice of operation: "))
    if o==1:
        sum=n1+n2
        print("required sum is:",sum)
    elif o==2:
        diff=n1-n2
        print("required difference is:",diff)
    elif o==3:
        product=n1*n2
        print("required product is:",product)
    elif o==4:
        if n2==0:
            print("divsion with zero is not possible")
        else:
            div=n1/n2
            print("required answer after division is:",div)
    elif o==5:
        if n2==0:
            print("floor divsion with zero is not possible")
        else:
            fd=n1//n2
            print("required answer after floor division is:",fd)
    elif o==6:
        if n2==0:
            print("divsion with zero is not possible therefore cannot get remainder")
        else:
            r=n1%n2
            print("required remainder is:",r)
    elif o==7:
        print("exiting calculator")
        break
    else: print("invalid choice")

  


        

    
```
